```python
dict_by_studyid = defaultdict(list)
f = open('mimiccxr_train_with_count.csv', 'r')
rdr = csv.reader(f)
for i, line in enumerate(tqdm(rdr)):
    dicom_id, subject_id, study_id, ViewPosition, count = line
    key = 's' + study_id
    dict_by_studyid[key].append(line)
f.close()

dict_by_studyid
```

```python
f = open('mdvl_mimiccxr_train.csv', 'w')
wr = csv.writer(f)

for info_dict in data:
    studyid = info_dict['id']
    if studyid in dict_by_studyid.keys():
        lists = dict_by_studyid[studyid]
        for rowlist in lists:
            wr.writerow(rowlist)        
    else:
        print(f'not in dict_by_studyid: {studyid}')

f.close()
```
